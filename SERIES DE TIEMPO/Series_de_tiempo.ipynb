```python
# Importamos las bibliotecas necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Definimos la ruta del archivo
ruta_1 = "C:/Users/juanj/Workspace/business_Intelligence/SERIES DE TIEMPO/Competidor1.csv"

# Leemos el archivo csv
Competitor1 = pd.read_csv(ruta_1)

# Convertimos la columna 'Product1' en una serie de tiempo
ts_Competidor1 = pd.Series(Competitor1['Product1'].values, 
                           index=pd.date_range('2019-01-01', periods=len(Competitor1), freq='W'))

# Graficamos la serie de tiempo
ts_Competidor1.plot()

# Creamos el modelo ARIMA
model = ARIMA(ts_Competidor1, order=(1,0,0))

# Ajustamos el modelo
ARIMA_Competidor1 = model.fit()

# Imprimimos el resumen del modelo
print(ARIMA_Competidor1.summary())

# Graficamos los valores reales vs los pronosticados por el modelo
plt.figure()
plt.plot(ts_Competidor1, color='blue')
plt.plot(ARIMA_Competidor1.fittedvalues, color='red')
plt.title('Real vs Pronstico')

# Hacemos un pronstico para los prximos 25 periodos
Forecast_Competidor1 = ARIMA_Competidor1.forecast(steps=25)

# Imprimimos el pronstico
print(Forecast_Competidor1)

# Graficamos el pronstico
plt.figure()
plt.plot(Forecast_Competidor1[0])
plt.title('Pronstico Competidor1')
plt.ylabel('Volumen Competidor1')
```
