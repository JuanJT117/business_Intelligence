```python
#fusionamos las columnas
tabla = df_copy[["AGENT.ID","ClientID"]].merge(df_copy[["AGENT.ID","ClientID"]], on = "ClientID", how = "inner", validate = "m:m" ).drop_duplicates()
tabla
```

```python
#eliminamos los agentes X y Y que no comparten cliente
tabla = tabla.loc[tabla["AGENT.ID_x"]!= tabla["AGENT.ID_y"]]
tabla
```

```python
tabla.groupby(["AGENT.ID_x","AGENT.ID_y"]).agg({"ClientID":"nunique"}).reset_index().drop_duplicates()
```
